```r
mean_mpg <- mean(mtcars$mpg)                
trimmed_mean_mpg <- mean(mtcars$mpg, trim = 0.1)  
weighted_mean_mpg <- weighted.mean(mtcars$mpg, w = mtcars$wt) 
```

```r
quantiles_mpg <- quantile(mtcars$mpg)        
deciles_mpg <- quantile(mtcars$mpg, probs = seq(0, 1, by = 0.1)) 
percentiles_mpg <- quantile(mtcars$mpg, probs = seq(0, 1, by = 0.01)) 

```

```r

data(mtcars)


median_mpg <- median(mtcars$mpg)


print(median_mpg)


```

```r
list(mean_mpg = mean_mpg, trimmed_mean = trimmed_mean_mpg, weighted_mean = weighted_mean_mpg,
     median = median_mpg, quantiles = quantiles_mpg, deciles = deciles_mpg, percentiles = percentiles_mpg)
```

```r
range_mpg <- range(mtcars$mpg)

iqr_mpg <- IQR(mtcars$mpg)

interdecile_range_mpg <- deciles_mpg[10] - deciles_mpg[2]
```

```r
var_mpg <- var(mtcars$mpg)         # Variance
sd_mpg <- sd(mtcars$mpg)  
```

```r

skewness_mpg <- skewness(mtcars$mpg)
kurtosis_mpg <- kurtosis(mtcars$mpg)

print(paste("Skewness:", skewness_mpg))
print(paste("Kurtosis:", kurtosis_mpg))
```

```r
skewness_mpg <- skewness(mtcars$mpg)
kurtosis_mpg <- kurtosis(mtcars$mpg)
```

```r
list(range = range_mpg, IQR = iqr_mpg, interdecile_range = interdecile_range_mpg,
     variance = var_mpg, sd = sd_mpg, skewness = skewness_mpg, kurtosis = kurtosis_mpg)
```

```r
frequency_mpg <- table(cut(mtcars$mpg, breaks = 5))

hist(mtcars$mpg, main = "Histogram of MPG", xlab = "Miles per Gallon", col = "blue", freq = TRUE)
```

```r
rel_freq_mpg <- prop.table(frequency_mpg)


cum_freq_mpg <- cumsum(frequency_mpg)
```

```r
list(frequency = frequency_mpg, relative_frequency = rel_freq_mpg, cumulative_frequency = cum_freq_mpg)
```
